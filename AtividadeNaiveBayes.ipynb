```python
previsores = data.iloc[:,2:5].values
classe = data.iloc[:,5].values
```

```python
qtd_rest, qtd_walking, qtd_running = 0, 0, 0
kind = np.array(classe)
for i in range(0, len(kind)):
    if kind[i] == 'rest':
        qtd_rest += 1
    if kind[i] == 'walking':
        qtd_walking += 1
    if kind[i] == 'running':
        qtd_running += 1
prob_rest = qtd_rest/len(kind)
prob_walking = qtd_walking/len(kind)
prob_running = qtd_running/len(kind)
print(f'Probabilidade da classe Rest: {prob_rest}')
print(f'Probabilidade da classe Walking: {prob_walking}')
print(f'Probabilidade da classe Running: {prob_running}')
```

```python
previsores = data.iloc[:,2:6].values
#Função que calcula a probabilidade de determinado valor de um atributo por cada classe
def atributo_p_classe(atributo, classe, posicao):
    qtd = 0
    for i in range(0, len(previsores)):
        linha = previsores[i]
        if linha[posicao] == atributo and linha[3] == classe:
            qtd += 1
    return qtd
```

```python
#Probabilidade do atributo "diet" pelas classes

#Probabilidade de "low fat" pela classe "rest"
prob_lowfat_rest = atributo_p_classe('low fat','rest',0)/qtd_rest
print(prob_lowfat_rest)
#Probabilidade de "no fat" pela classe "rest"
prob_nofat_rest = atributo_p_classe('no fat','rest',0)/qtd_rest
print(prob_nofat_rest)
#Probabilidade de "low fat" pela classe "walking"
prob_lowfat_walking = atributo_p_classe('low fat','walking',0)/qtd_walking
print(prob_lowfat_walking)
#Probabilidade de "no fat" pela classe "walking"
prob_nofat_walking = atributo_p_classe('no fat','walking',0)/qtd_walking
print(prob_nofat_walking)
#Probabilidade de "low fat" pela classe "running"
prob_lowfat_running = atributo_p_classe('low fat','running',0)/qtd_running
print(prob_lowfat_running)
#Probabilidade de "no fat" pela classe "running"
prob_nofat_running = atributo_p_classe('no fat','running',0)/qtd_running
print(prob_nofat_running)
```

```python
# Função que calcula o atributo pulse por cada classe

def pulse_p_classe(classe, posicao):
    qtd = 0
    for i in range(0, len(previsores)):
        linha = previsores[i]
        if linha[3] == classe:
            qtd += linha[1]
    return qtd
```

```python
# Média do atributo pulse por classe

media_rest = pulse_p_classe('rest', 2)/qtd_rest
print(media_rest)
media_walking = pulse_p_classe('walking', 2)/qtd_walking
print(media_walking)
media_running = pulse_p_classe('running', 2)/qtd_running
print(media_running)
```

```python
#Função que calcula o desvio padrão
def calculaDesvioPadrao(classe, media, qtd_classe):
    dif = 0
    m = media
    for i in range(0, len(previsores)):
        linha = previsores[i]
        if linha[3] == classe:
            dif += math.pow(linha[1] - media, 2)
    razao = dif/(qtd_classe-1)
    return math.sqrt(razao)
```

```python
# Desvio padrão de cada classe

dv_rest = calculaDesvioPadrao('rest', media_rest, qtd_rest)
print(dv_rest)
dv_walking = calculaDesvioPadrao('walking', media_walking, qtd_walking)
print(dv_walking)
dv_running = calculaDesvioPadrao('running', media_running, qtd_running)
print(dv_running)
```

```python
#Função que calcula a gaussiana
def gaussian(x, media, desvio):
    exp = math.exp(-(math.pow(x-media, 2)/(2*math.pow(desvio, 2))))
    return (1/(math.sqrt(2*math.pi)*desvio))*exp
```

# Vamos adcionar uma nova instância ao nosso conjunto de dados com os valores diet = low fat, pulse = 90 e time = 1 min e classificá-la.

```python
# Gaussiana de cada classe para um novo exemplo:
# pulse = 90

gaussian_rest = gaussian(90, media_rest, dv_rest)
print(gaussian_rest)
gaussian_walking = gaussian(90, media_walking, dv_walking)
print(gaussian_walking)
gaussian_running = gaussian(90, media_running, dv_running)
print(gaussian_running)
```

```python
def probabilidade_final(classe, diet_classe, gaussian_classe, time_classe):
    return (classe * diet_classe * gaussian_classe * time_classe)
    

prod_rest = probabilidade_final(prob_rest, prob_lowfat_rest, gaussian_rest, prob_1_rest)
print(f'A probabilidade de ser da classe rest é de {prod_rest}')
prod_walking = probabilidade_final(prob_walking, prob_lowfat_walking, gaussian_walking, prob_1_walking)
print(f'A probabilidade de ser da classe walking é de {prod_walking}')
prod_running = probabilidade_final(prob_running, prob_lowfat_running, gaussian_running, prob_1_running)
print(f'A probabilidade de ser da classe running é de {prod_running}\n')

if prod_rest > prod_walking and prod_rest > prod_running:
    print('A nova instância pertence a classe REST')
if prod_walking > prod_rest and prod_walking > prod_running:
    print('A nova instância pertence a classe WALKING')
if prod_running > prod_rest and prod_running > prod_walking:
    print('A nova instância pertence a classe RUNNING')
```
